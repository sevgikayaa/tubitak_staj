```python
from lightkurve import search_targetpixelfile

pixelfile = search_targetpixelfile("KIC 9832227", quarter=16).download(quality_bitmask='hardest');
```

```python
import matplotlib.pyplot as plt
pixelfile.plot(frame=1000)
plt.show()
```

```python
pixelFile = search_targetpixelfile('KIC 9832227', quarter=4).download()
```

```python
lc = pixelFile.to_lightcurve(aperture_mask=pixelFile.pipeline_mask)
```

```python
lc.plot()
plt.show()
```

```python
flat_lc = lc.flatten(window_length=401)
flat_lc.plot()
plt.show()
```

```python
import numpy as np
periodogram = flat_lc.to_periodogram(
    method="bls", 
    period=np.arange(1, 10, 0.001),
     # hata verdii iin 0.3 ten kk bir deer girildi
)
periodogram.plot()
plt.show()
```

```python
best_fit_period = periodogram.period_at_max_power
print('Best fit period: {:.5f}'.format(best_fit_period))
```

```python
folded_lc = flat_lc.fold(period=1.145)
folded_lc.plot()
plt.show()
```

```python
binned_lc = folded_lc.bin(time_bin_size=0.001)
binned_lc.plot()
plt.show()
```
